```python
import itertools
import pandas as pd
import pathlib
import re

from utils import get_df, get_index_mean, get_index_std, get_column, minimums, colors, descriptors

regioes = ['Norte', 'Nordeste', 'Sul', 'Sudeste', 'Centro-Oeste']

dfs = {'%s+%s+%s' % (regiao, minimum, color): get_df(color) for color in colors for minimum in minimums for regiao in regioes}
dataset_name = 'regioes'
os.makedirs(dataset_name, exist_ok=True)
```

```python
def parse_folder_name(folder_name):
    pat = r'clf\=(.*?)\+len\=(.*?)\+ex\=(.*?)\+ft\=(.*?)\+c\=(.*?)\+dt\=(.*?)\+r\=(.*?)\+m\=(.*?)$'
    finds = re.findall(pat, folder_name)
    finds = list(itertools.chain(*finds))
    if len(finds) > 0:
        return {'classifier': finds[0], 'len': finds[1], 'extractor': finds[2], 'features': finds[3], 'color': finds[4],
                'dataset': finds[5], 'regiao': finds[6], 'minimum': finds[7]}
    return {}

def get_df_sheet(infos, minimum, regiao):
    return '%s+%s+%s' % (regiao, str(minimum), infos['color'])

for p in pathlib.Path('/media/xandao/d3b6c029-9e4c-4225-a0df-da4f12bf464c/regions').rglob('mean+f1+sum.csv'):
    for minimum in minimums:
        for c in colors:
            for regiao in regioes:
                infos = parse_folder_name(p.parent.parent.parent.name)
                if len(infos) > 0 and (str(infos['minimum']) == str(minimum) and str(infos['features']) == str(descriptors[infos['extractor']][0]) and c == str(infos['color']) and regiao == str(infos['regiao'])):
                    df = pd.read_csv(p, index_col=0, header=None, sep=';')
                    dfs[get_df_sheet(infos, minimum, regiao)].loc[get_index_mean(infos), get_column(infos)] = \
                        df.loc['mean_f1'].values[0]
                    dfs[get_df_sheet(infos, minimum, regiao)].loc[get_index_std(infos), get_column(infos)] = \
                        df.loc['std_f1'].values[0]
```

```python
for k, v in dfs.items():
    v.sort_index(inplace=True)
    display(k, v)
    v.to_csv(os.path.join(dataset_name, '%s.csv') % k, sep=';', float_format=',', quoting=2, na_rep=0)
```
